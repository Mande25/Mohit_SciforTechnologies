```python
import pandas as pd

# Create a dummy dataset with 20 rows and 3 columns
data5 = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Alice', 'David', 'Alice', 'Bob', 'Eva', 'Frank', 'Grace', 'Harry', 'Ivy', 'Jack', 'Alice', 'Bob', 'Charlie', 'Alice', 'David', 'Alice', 'Bob'],
    'Age': [25, 30, 22, 25, 35, 25, 30, 28, 40, 35, 22, 33, 27, 25, 30, 22, 25, 35, 25, 30],
    'City': ['New York', 'Los Angeles', 'Chicago', 'New York', 'San Francisco', 'New York', 'Los Angeles', 'San Francisco', 'Chicago', 'New York', 'Los Angeles', 'Chicago', 'San Francisco', 'New York', 'Los Angeles', 'Chicago', 'New York', 'San Francisco', 'New York', 'Los Angeles'],
}

df5 = pd.DataFrame(data5)

# Display the dummy dataset
print(df5)
```

```python
# Check for duplicates based on the 'Name' column
duplicates_name = df5.duplicated(subset=['Name'])

# Print the DataFrame with a new column indicating duplicate status for the specified column
df5['IsDuplicatedName'] = duplicates_name

# Display the DataFrame
print(df5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count the occurrences of each name
name_counts = df5['Name'].value_counts()

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=name_counts.index, y=name_counts.values, palette="viridis")
plt.title('Duplicate Values in Name Column')
plt.xlabel('Name')
plt.ylabel('Count')

plt.show()
```

```python
# Remove duplicates based on the 'Name' column
df5_no_duplicates = df5.drop_duplicates(subset=['Name'])

# Display the DataFrame without duplicates
print(df5_no_duplicates)
```
